=> Q24. item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장
* 데이터 프레임 에서 컬럼의 데이터 문자열 일부값만 추출 .str

```python
df['new_price'] = list(map(lambda x : float(x[1:]), df['item_price']))
df['new_price']
```

```python
df['new_price'] = df['item_price'].str[1:].astype('float')
df['new_price']
```

Q25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수 구하기

```python
len(df.loc[df['new_price'] <= 5])
```

Q26. item_name명이 Chicken Salad Bowl인 데이터 프레임을 추출하고 index 값을 초기화

```python
ans = df[df['item_name'] == "Chicken Salad Bowl"].reset_index(drop=True)
ans
```

Q27. new_price 값이 9 이하이고, item_name 값이 "Chicken Salad Bowl인 데이터 프레임 추출

```python
ans = df[df.new_price <= 9][df.item_name == "Chicken Salad Bowl"]
ans.head()
```

```python
df.loc[(df.item_name =='Chicken Salad Bowl') & (df.new_price <= 9)].head()
```
